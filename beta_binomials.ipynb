```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt

if __name__ == "__main__":
    # Creat a list of the number of coin tosses ("Bernoulli trials")
    number_of_trials = [0, 2, 10, 20, 50, 500]

    # Conduct 500 coin tosses and output into a list of 0s and 1s
    # where 0 represents a tail and 1 represents a head
    data = stats.bernoulli.rvs(.5, size=number_of_trials[-1])

    x = np.linspace(0, 1, 100)

    # Loop over the number of tirals
    # add more coin toss data. For each new set of data
    # update current prior belief to be new posterior
    # carried out using what is known as beta-binonomial model

    for i, N in enumerate(number_of_trials):
      #Accumulate the total number of heads
      heads = data[:N].sum()

      #create axes subplot for each update
        # create axes subplot for each update
      ax = plt.subplot(len(number_of_trials) // 2, 2, i + 1)
      ax.set_title("%s trials, %s heads" % (N, heads))


        # set y-axis limit for the first subplot
      if i == 0:
       plt.ylim([0.0, 2.0])
       plt.setp(ax.get_yticklabels(), visible=False)


       # Create and plot a Beta distribution to represent the
# posterior belief in fairness of the coin.
      y = stats.beta.pdf(x, 1 + heads, 1 + N - heads)
      plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
      plt.fill_between(x, 0, y, color="#aaaadd", alpha=0.5)

        # hide y-axis tick labels
      ax.set_yticklabels([])

# show the plots
plt.tight_layout()
plt.show()


```
